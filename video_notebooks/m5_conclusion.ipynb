### Make Changes in a DataFrame

```python
df = pd.DataFrame({'first name':['Ron', 'Linden', 'Kevin', 'Marco'], 'score':[90, 95, 85, 93], 'grade':['A-', 'A', 'B', 'A-']})
df
```

```python
df[df.score>=93]
```

```python
df[df.score>=93]['grade']
```

```python
df[df['first name'] == 'Linden']
```

```python
#set grade to A
df[df.score>=93]['grade']='A' 
```

```python
df
```

```python
df.loc[df.score>=93, 'grade']='A'
df
```
