# Problem 1: Read a table from sqlite db to a DataFrame

- The sqlite database file name is 'm8-high.db'.
- Establish a connection `con` to the database.
- Use pandas read_sql function with database connect represented by `con` to load all data from the table `Courses` and load the result to a DataFrame df_courses.
- Display the first 5 rows of df_courses.


```python
# Read table from db
import sqlite3 as sql
import pandas as pd

con = sql.connect('m8-high.db')
df_courses = pd.read_sql('Select * from Courses', con)
df_courses.head()
```

## Problem 2: Select 500-level courses 

- Use Pandas read_sql() to load all courses with course_num greater or equal to 500 in `Courses` table to a DataFrame `df_500`.
- Display the first 5 rows of `df_500`


```python
df_500 = pd.read_sql("select * from Courses where course_num>=500", con)
df_500.head()
```

# Problem 3: Select courses with "data" in course name

- Use Pandas read_sql() to load all courses that have word "data" in `course_name` column. Save the result to DataFrame `df_data`.
- Display the `df_data`.

**Hint:** Use SQL LIKE operator.

```python
df_data = pd.read_sql("select * from Courses where course_name like '%data%'", con)
df_data
```
