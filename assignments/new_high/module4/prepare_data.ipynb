```python
import pandas as pd
df = pd.read_csv('gdp_cap_all.csv')
df.head()
```

```python
df['Country Name'].unique()
```

```python
df1 = df[df['Country Name'].isin(['China', 'United States', 'United Kingdom', 'Russian Federation', 'France', 'India', 'Japan', 'Germany', 'Australia', 'South Africa'])]
df1 = df1[['Country Name', '2016', '2017', '2018', '2019']]
df1['2016'] = df1['2016'].astype(int)
df1['2017'] = df1['2017'].astype(int)
df1['2018'] = df1['2018'].astype(int)
df1['2019'] = df1['2019'].astype(int)
df1 = df1.reset_index(drop=True)
df1.to_csv('gdp_cap_10.csv', index=False)
```
