```python
import pandas as pd
df = pd.read_csv('wealth_and_health.csv')
df.head()
```

```python
df_2007 = df[df.year==2007]
df_2007 = df_2007.reset_index(drop=True)
df_2007.head()
```

```python
df_2007.pivot_table(index='continent', values=['life_exp', 'gdp_cap'], aggfunc='mean')
```

```python
#convert 5973.02 to $5,973
def convert_currency(amount):
    s = str(int(amount))
    i = 3
    while i < len(s):
        s = s[:-i]+','+s[-i:]
        i += 4
    return '$'+s
convert_currency(1234567890.0001)
```

```python
df_2007.gdp_cap = df_2007.gdp_cap.apply(convert_currency)
df_2007.head()
```

```python
for i in [2,5,13,83,131]:
    df_2007.loc[i, 'gdp_cap'] = None
for i in [3,54,111,120]:
    df_2007.loc[i, 'life_exp'] = None
```
