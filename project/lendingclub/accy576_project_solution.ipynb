### Task 19: Loan Return

Calculate the return of a loan portfolio.

Calculation of loan return is very complicated since the loan is paid by monthly installments. In this project, we simplify the calculation by using the total payment and funded amount. For charged off loans, total payment includes post charge off recoveries. So we can use following formula to calculate the total return:

$TotalReturn = \frac{Total Payment + Recoveries}{Funded Amount} - 1$

The overall return doesn't reflect loan profitbility since loans have different terms. It's more accurate to compare annual returns. There are only two terms in the dataset, 36 months and 60 months. The formula to calcuate annual return is:

$Annualized Return = (1+Total Return)^{(1/years)} - 1$. 

For example, if overall return of the 36 month loan portfolio is 10%, then annualized return = `(1 + 0.1)**(1/3) - 1 = 0.032`.

To finish this task, do:
- Strip extra white space in term column, ie. convert ' 36 months' to '36 months'.
- Create two DataFrames, one with all 36 months loan and one with all 60 months loan.
- For each DataFrame:
    - Get total payment of all loans that includes recoveries of charged off loans
    - Get total funded amount of all loans.
    - Calculate overal return.
    - Calculate annualized return. Assign annualized return of 36 month and 60 month loans to **annual_return_36 and annual_return_60**
- Discuss the result briefly

#### Your Code

```python
loan_df.term = loan_df.term.str.strip()
df_36 = loan_df[loan_df.term=="36 months"]
return_36 = (df_36.total_pymnt.sum() + df_36.recoveries.sum())/df_36.funded_amnt.sum()-1
annual_return_36 = (1+return_36)**(1/3)-1
annual_return_36
```

```python
df_60 = loan_df[loan_df.term=="60 months"]
return_60 = (df_60.total_pymnt.sum() + df_60.recoveries.sum())/df_60.funded_amnt.sum()-1
return_60
annual_return_60 = (1+return_60)**(1/5)-1
annual_return_60
```

#### Your Discussion
60 month loan has higer annualized return.

### Task 20: Estimate Interest Rate

Build an OLS regression model to predict loan interest rate.

- Choose interest rate as dependent variable
- Choose at least 4 columns as independent variables which should have at least one categorical features and at least one continuous features.
- Use statsmodels to perform OLS linear regression.
- Display and discuss the regression result.
 - The quality of the regression model(r-squared score)
 - The significancy of each coefficient.

#### Your Code

```python
loan_df.info()
```

```python
import statsmodels.formula.api as smf
formula = 'int_rate ~ loan_amnt + annual_inc + C(term) + C(grade)'
model = smf.ols(formula, data=loan_df)
result = model.fit()
result.summary()
```

#### Your Discussion
